```python
with open('weekly_patterns.csv','w',newline='') as csvfile:
    fieldnames = ['#Case','Initial_Investment_amount','Buy_Weekday','Buy_at','Sell_Weekday','Sell_at','From Date','To Date','Total_Time_Period','rollOverpPattern','Total Investment','Total profit','% Gain']
    thewriter= csv.DictWriter(csvfile, fieldnames=fieldnames)
    thewriter.writeheader()
count_case=0
```

```python
initial_investment = 100
initial_invest_amount=initial_investment
buy_weekday = 'Monday'
buy_at = 'Open'
sell_weekday = 'Monday'
sell_at = 'Close'
total_investment = 0
total_profit = 0
roll_Over="Yes"
df_set=df_2015
weekdays=['Monday','Tuesday','Wednesday','Thursday','Friday']
row=df_set.index
count=len(row)
#count_case=0
with open('weekly_patterns.csv','a',newline='') as csvfile:
    fieldnames = ['#Case','Initial_Investment_amount','Buy_Weekday','Buy_at','Sell_Weekday','Sell_at','From Date','To Date','Total_Time_Period','rollOverpPattern','Total Investment','Total profit','% Gain']
    thewriter= csv.DictWriter(csvfile, fieldnames=fieldnames)
    #thewriter.writeheader()
    for day in weekdays:
        buy_weekday = day
        sell_weekday= day
        for data_row in df_set.index:
            if df_set['weekday'][data_row] == buy_weekday:
                stock_amount = initial_investment/df_set[buy_at][data_row]
                stock_selling_amount = df_set[sell_at][data_row]*stock_amount
                stock_profit = stock_selling_amount-initial_investment
                total_profit+=stock_profit
                total_investment+=initial_invest_amount
                if roll_Over=="Yes":
                    initial_investment=initial_investment+stock_profit

        title="Every {0} Purchase at {1}, and sell at {2} with $100 initial capital"
        print(title.format(buy_weekday,buy_at,sell_at))
        print("Total Investment :",total_investment)
        print("Total Return :",total_investment+total_profit)
        count_case+=1
        percentage = (total_profit/total_investment)*100
        print("profit:",round(total_profit, 2))
        print("Percentage Gain:",round(percentage, 2))
        print("\n")
        time_difference=str(df_set.iloc[count-1,0] - df_set.iloc[0,0]).split()
        Total_Time_Period=time_difference[0]+" "+time_difference[1]
        thewriter.writerow({'#Case':count_case,'Initial_Investment_amount':initial_invest_amount,'Buy_Weekday':buy_weekday,'Buy_at':buy_at,'Sell_Weekday':sell_weekday,'Sell_at':sell_at,'From Date':df_set.iloc[0,0].date(),'To Date':df_set.iloc[count-1,0].date(),'Total_Time_Period':Total_Time_Period,'rollOverpPattern':roll_Over,'Total Investment':total_investment,'Total profit':round(total_profit, 2),'% Gain':round(percentage, 2)})
        total_investment = 0
        total_profit = 0
        initial_investment = 100
```
